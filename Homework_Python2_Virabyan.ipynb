```python
import numpy as np
start = time.time()
products = np.array(['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны'])
prices = np.array([20000, 3000, 45600, 5000], dtype="int64")
volume_sales = np.array([sum(sales_avto), sum(sales_ob), sum(sales_od), sum(sales_tel)])
revenue = np.zeros(4)
np.multiply(volume_sales, prices, out=revenue)
print('Доход от автомобилей HAVAL:', revenue[0])
print('Доход от обуви:', revenue[1])
print('Доход от одежды:', revenue[2])
print('Доход от телефонов:', revenue[3])
end = time.time() - start
print('Время выполнения:', end)
```

Как мы видим, у доходов продуктов во всех реализациях значения одинаковые, время у Numpy и Numpy с векторизацией намного лучше чем при классическом Python. Также стоит отметить, что почему-то время выполнения у Numpy.sum в варианте с векторизацией получается больше, чем при обычном sum.

```python
import numpy as np
def mergeSort_numpy(myList):
    if len(myList) > 1:
        
        mid = len(myList) // 2
        left = np.array(myList[:mid])
        
        right = np.array(myList[mid:])
        

        mergeSort_numpy(left)
        
        mergeSort_numpy(right)
         
        i = 0
        j = 0
        
        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
            
              myList[k] = left[i]
              
              i += 1
            else:
                myList[k] = right[j]
                j += 1
            
            k += 1
        while i < len(left):
            myList[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            myList[k]=right[j]
            j += 1
            k += 1

myList = np.array([54, 26, 93, 17, 77, 31, 44, 55, 20, 26, 93, 17, 77, 31, 44, 55, 20, 26, 93, 17, 77, 31, 44, 55, 20])
mergeSort_numpy(myList)
print(myList)
```

Реализация mergesort на классическом Python

```python
def mergeSort(myList):
    if len(myList) > 1:
        mid = len(myList) // 2
        left = myList[:mid]
        right = myList[mid:]

        mergeSort(left)
        mergeSort(right)

        i = 0
        j = 0

        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
              myList[k] = left[i]
              
              i += 1
            else:
                myList[k] = right[j]
                j += 1
          
            k += 1

        while i < len(left):
            myList[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            myList[k]=right[j]
            j += 1
            k += 1

myList = [54, 26, 93, 17, 77, 31, 44, 55, 20, 26, 93, 17, 77, 31, 44, 55, 20, 26, 93, 17, 77, 31, 44, 55, 20]
mergeSort(myList)
print(myList)
```

```python
import numpy as np
dtype = [('movie', 'U50'), ('type', 'U50'), ('word', 'U50'), ('minutes_in', 'f8')]
data = np.genfromtxt('/pap/tarantino.csv', delimiter=',', skip_header=1, dtype=dtype)

print(np.sum(data["type"] == "word")) # определяем число проклятий
print(np.sum(data["type"] == "death")) # определяем число смертей

print("Определяем соотношения проклятий/смертей по фильмам")
for i in np.unique(data["movie"]):
    print (i, np.sum((data["type"] == "word") & (data["movie"] == i))/np.sum((data["type"] == "death") & (data["movie"] == i)))

print("Определяем частоту употреблений проклятий")
for i in np.unique(data[data["type"] == "word"]["word"]):
    print (i, np.sum(data["word"] == i))
```
